```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11  # Gravitational constant, in m^3 kg^-1 s^-2
M_star = 1.989e30  # Mass of the star (Sun), in kg
orbital_radius = 1.496e11  # Average distance from Earth to Sun, in meters
orbital_period = 365.25 * 24 * 3600  # Orbital period in seconds (1 year)

# Calculate orbital velocity for circular orbit
velocity = np.sqrt(G * M_star / orbital_radius)  # in m/s

# Time array for simulation
total_time = orbital_period  # simulate for one full orbit
dt = 24 * 3600  # time step: 1 day in seconds
times = np.arange(0, total_time, dt)

# Initialize position arrays
x_positions = []
y_positions = []

# Initial position and velocity
x = orbital_radius
y = 0
vx = 0
vy = velocity

for t in times:
    # Append current position
    x_positions.append(x)
    y_positions.append(y)
    
    
    # Calculate acceleration towards the star
    r = np.sqrt(x**2 + y**2)
    a = G * M_star / r**2
    
    # Components of acceleration
    ax = -a * (x / r)
    ay = -a * (y / r)
    
    # Update velocities
    vx += ax * dt
    vy += ay * dt
    
    # Update positions
    x += vx * dt
    y += vy * dt

# Plot the orbit
plt.figure(figsize=(8,8))
plt.plot(x_positions, y_positions, label='Orbital Path')
plt.scatter([0], [0], color='yellow', label='Star (Sun)', s=300)
plt.title('Simulated Planetary Orbit')
plt.xlabel('X Position (meters)')
plt.ylabel('Y Position (meters)')
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
